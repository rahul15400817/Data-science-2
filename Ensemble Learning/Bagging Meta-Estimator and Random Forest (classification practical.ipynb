```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000,noise=0.2)
```

```python
d={'x1':x[:,0],'x2':x[:,1],'y':y}
```

```python
df=pd.DataFrame(d)
```

```python
df.head(3)
```

```python
x_a=df.iloc[:,:-1]
y_a=df['y']
```

```python
sns.scatterplot(x='x1',y='x2',data=df,hue='y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_a,y_a,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
gc=BaggingClassifier(base_estimator=SVC(),n_estimators=20)
gc.fit(x_train,y_train)
```

```python
gc.score(x_train,y_train)*100,gc.score(x_test,y_test)*100
```

```python
rc=RandomForestClassifier(n_estimators=20)
rc.fit(x_train,y_train)
```

```python
rc.score(x_train,y_train)*100, rc.score(x_test,y_test)*100
```
