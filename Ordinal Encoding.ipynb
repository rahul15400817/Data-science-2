```python
import pandas as pd
```

```python
df=pd.DataFrame({'Size':['s','m','l','xl','s','xl','l','s','l','xl','m']})
df.head()
```

```python
ord_data=[['s','m','l','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
```

```python
oe.fit(df[['Size']])
```

```python
df['size_en']=oe.transform(df[['Size']])
```

```python
df
```

```python
#Ordinal Encoding using map function
ord_data1={'s':3,'m':5,'l':6,'xl':9}
```

```python
df['Size_en_map']=df['Size'].map(ord_data1)
```
