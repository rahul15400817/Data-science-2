```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
placed=pd.read_csv(r"C:\Users\rku68\OneDrive\Desktop\csv\placement_dataset.csv")
```

```python
placed.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='cgpa',y='iq',data=placed,hue='placement')
plt.show()
```

```python
sns.kdeplot(data=placed['iq'])
plt.show()
```

```python
x=placed.iloc[:,:-1]
y=placed['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.predict([[5.9,106]])
```

```python
gnb.score(x_test,y_test)*100, gnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
bnb=BernoulliNB()
```

```python
bnb.fit(x_test,y_test)
```

```python
bnb.score(x_test,y_test)*100, bnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100 ,mnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```
